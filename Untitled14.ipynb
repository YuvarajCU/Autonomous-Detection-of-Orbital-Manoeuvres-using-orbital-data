```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.svm import OneClassSVM
from sklearn.preprocessing import StandardScaler

# Load and preprocess data
url = 'https://drive.google.com/uc?id=1BvQRvWXdbpXpOaSahEMovATmKQuQilQa'
data = pd.read_csv(url, parse_dates=['Datetime'])

# Assuming 'SMA' is your semi-major axis column
X = data[['SMA']].values

# Standardize the data
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Train the One-Class SVM model
model = OneClassSVM(nu=0.01)  # Adjust nu based on expected outlier/anomaly ratio
model.fit(X_scaled)

# Predict outliers/anomalies
y_pred = model.predict(X_scaled)

# Visualize anomalies
plt.figure(figsize=(14, 7))
plt.plot(data['Datetime'], data['SMA'], label='SMA')
plt.scatter(data['Datetime'][y_pred == -1], data['SMA'][y_pred == -1], color='red', label='Anomalies')
plt.xlabel('DateTime')
plt.ylabel('SMA')
plt.title('Detection of Potential Orbital Maneuvers (Anomalies) using One-Class SVM')
plt.legend()
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.neighbors import LocalOutlierFactor
from sklearn.preprocessing import StandardScaler

# Load and preprocess data
url = 'https://drive.google.com/uc?id=1BvQRvWXdbpXpOaSahEMovATmKQuQilQa'
data = pd.read_csv(url, parse_dates=['Datetime'])

# Assuming 'SMA' is your semi-major axis column
X = data[['SMA']].values

# Standardize the data
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Train the Local Outlier Factor (LOF) model
model = LocalOutlierFactor(contamination=0.01)  # Adjust contamination based on expected outlier/anomaly ratio
y_pred = model.fit_predict(X_scaled)

# Convert y_pred to binary labels where -1 indicates anomaly
is_anomaly = y_pred == -1

# Visualize anomalies
plt.figure(figsize=(14, 7))
plt.plot(data['Datetime'], data['SMA'], label='SMA')
plt.scatter(data['Datetime'][is_anomaly], data['SMA'][is_anomaly], color='red', label='Anomalies')
plt.xlabel('DateTime')
plt.ylabel('SMA')
plt.title('Detection of Potential Orbital Maneuvers (Anomalies) using Local Outlier Factor (LOF)')
plt.legend()
plt.show()
```

```python
from sklearn.mixture import GaussianMixture
from sklearn.preprocessing import StandardScaler

# Load and preprocess data
url = 'https://drive.google.com/uc?id=1BvQRvWXdbpXpOaSahEMovATmKQuQilQa'
data = pd.read_csv(url, parse_dates=['Datetime'])

# Assuming 'SMA' is your semi-major axis column
X = data[['SMA']].values

# Standardize the data
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Train the Gaussian Mixture Model (GMM)
gmm = GaussianMixture(n_components=2, random_state=42)
gmm.fit(X_scaled)

# Predict the probability of each point being an outlier
scores = gmm.score_samples(X_scaled)
threshold = np.percentile(scores, 5)  # Adjust threshold based on desired outlier ratio

# Identify anomalies
is_anomaly = scores < threshold

# Visualize anomalies
plt.figure(figsize=(14, 7))
plt.plot(data['Datetime'], data['SMA'], label='SMA')
plt.scatter(data['Datetime'][is_anomaly], data['SMA'][is_anomaly], color='red', label='Anomalies')
plt.xlabel('DateTime')
plt.ylabel('SMA')
plt.title('Detection of Potential Orbital Maneuvers (Anomalies) using Gaussian Mixture Model (GMM)')
plt.legend()
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.metrics import classification_report, accuracy_score, precision_score, recall_score, f1_score
from imblearn.over_sampling import SMOTE

# Load data
url = 'https://drive.google.com/uc?id=1BvQRvWXdbpXpOaSahEMovATmKQuQilQa'
data = pd.read_csv(url, parse_dates=['Datetime'])

# Data preprocessing
data['SMA_diff'] = data['SMA'].diff().fillna(0)
data['SMA_diff_abs'] = data['SMA_diff'].abs()
data['SMA_diff2'] = data['SMA_diff'].diff().fillna(0)  # Second difference
data['SMA_rolling_mean'] = data['SMA'].rolling(window=5).mean().fillna(data['SMA'].mean())
data['SMA_rolling_std'] = data['SMA'].rolling(window=5).std().fillna(data['SMA_diff'].std())

# Label the data
# Here we assume that the known maneuvers are labeled as:
# 0 - No maneuver
# 1 - Orbit raising
# 2 - Orbit lowering
# 3 - Drifting

def label_maneuvers(row, mean_diff, std_diff):
    if row['SMA_diff'] > mean_diff + 0.5 * std_diff:
        return 1  # Orbit raising
    elif row['SMA_diff'] < -(mean_diff + 0.5 * std_diff):
        return 2  # Orbit lowering
    elif abs(row['SMA_diff']) >= 0.01:
        return 3  # Drifting
    else:
        return 0  # No maneuver

mean_diff = data['SMA_diff_abs'].mean()
std_diff = data['SMA_diff_abs'].std()
data['Maneuver'] = data.apply(label_maneuvers, axis=1, mean_diff=mean_diff, std_diff=std_diff)

# Define features and target
features = ['SMA', 'SMA_diff', 'SMA_diff_abs', 'SMA_diff2', 'SMA_rolling_mean', 'SMA_rolling_std']
X = data[features].values
y = data['Maneuver'].values

# Apply SMOTE to balance the classes
smote = SMOTE(random_state=42)
X_resampled, y_resampled = smote.fit_resample(X, y)

# Split data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X_resampled, y_resampled, test_size=0.3, random_state=42)

# Initialize and train the Random Forest classifier
model = RandomForestClassifier(n_estimators=100, random_state=42, class_weight='balanced')
model.fit(X_train, y_train)

# Predict on the test set
y_pred = model.predict(X_test)

# Calculate classification report
print("Classification Report for RandomForestClassifier:")
print(classification_report(y_test, y_pred, target_names=['No Maneuver', 'Orbit Raising', 'Orbit Lowering', 'Drifting']))

# Calculate cross-validated scores
scores = cross_val_score(model, X_resampled, y_resampled, cv=5, scoring='f1_weighted')
print("Cross-validated F1 scores for RandomForestClassifier:")
print(scores)
print("Mean cross-validated F1 score:", np.mean(scores))

# Calculate additional metrics
precision = precision_score(y_test, y_pred, average='weighted')
recall = recall_score(y_test, y_pred, average='weighted')
f1 = f1_score(y_test, y_pred, average='weighted')

print("Weighted Precision of RandomForestClassifier:", precision)
print("Weighted Recall of RandomForestClassifier:", recall)
print("Weighted F1 Score of RandomForestClassifier:", f1)

# Predict anomalies on entire dataset
y_pred_all = model.predict(X)

# Extract dates with detected anomalies by RandomForestClassifier
anomaly_indices_rf = np.where(y_pred_all != 0)[0]
anomaly_dates_rf = data.iloc[anomaly_indices_rf]['Datetime']
maneuver_types = y_pred_all[anomaly_indices_rf]

# Plotting to visualize anomalies detected by RandomForestClassifier
plt.figure(figsize=(14, 7))
plt.plot(data['Datetime'], data['SMA'], label='Actual SMA')

# Plot detected maneuvers based on predicted SMA changes
for idx, maneuver in zip(anomaly_indices_rf, maneuver_types):
    if maneuver == 1:
        plt.scatter(data.iloc[idx]['Datetime'], data.iloc[idx]['SMA'], label='Orbit Raising', color='green', marker='^')
    elif maneuver == 2:
        plt.scatter(data.iloc[idx]['Datetime'], data.iloc[idx]['SMA'], label='Orbit Lowering', color='red', marker='v')
    elif maneuver == 3:
        plt.scatter(data.iloc[idx]['Datetime'], data.iloc[idx]['SMA'], label='Drifting', color='blue', marker='o')

# Display legend with unique labels
handles, labels = plt.gca().get_legend_handles_labels()
unique_labels = list(set(labels))  # Get unique labels
plt.legend(handles=[handles[labels.index(label)] for label in unique_labels], labels=unique_labels,
           bbox_to_anchor=(1.05, 1), loc='upper left')

plt.xlabel('DateTime')
plt.ylabel('SMA')
plt.title('Detected Orbital Maneuvers using Random Forest Classifier')
plt.show()
```
